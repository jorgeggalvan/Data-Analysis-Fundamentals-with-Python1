#### Ejercicio 5.1
---

> Dataset a utilizar: `amzn_stock.csv`

**5.1A:** Representa visualmente la evoluci√≥n temporal de la cotizaci√≥n 'Close' de las acciones de Amazon desde el 2020.

**5.1B:** Visualiza la evoluci√≥n de las acciones de Amazon a lo largo del tiempo, mostrando dos subgr√°ficos: uno para el precio de cierre y otro para el volumen de transacciones.

##### Ejercicio 5.1A

```python
# Importamos Pandas y Matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Utilizamos el "magic command" de Jupyter para desplegar gr√°ficos
%matplotlib inline
```

```python
# Importamos el DataFrame con las cotizaciones de Amazon
df_amzn = pd.read_csv('./data/amzn_stock.csv')
df_amzn.head()
```

```python
# Filtramos para seleccionar las filas con una fecha posterior al 01/01/2020
df_amzn_filtered = df_amzn[df_amzn['Date'] > '2020-01-01']
```

```python
# Creamos un gr√°fico de l√≠neas con la evoluci√≥n temporal del 'Close'
df_amzn_filtered.plot(kind='line', x='Date', y='Close', ylim=0,
                      title='Amazon Stock Price Evolution', color='green', figsize=(10,5))
plt.show()
```

##### Ejercicio 5.1B

```python
# Creamos dos subgr√°ficos de l√≠neas para visualizar la evoluci√≥n de 'Close' y 'Volume'
df_amzn.plot(kind='line', x='Date', y=['Close', 'Volume'], ylim=0,
                      title='Amazon Stock Evolution', color=['green', 'steelblue'], 
                      subplots=True, layout=(2,1), figsize=(10,5))
plt.show()
```

#### Ejercicio 5.2
---

> Dataset a utilizar: `imdb_movies.csv`

**5.2A:** Obt√©n los 10 actores con mayor cantidad de apariciones como protagonistas en pel√≠culas ('actor_1_name').

**5.2B:** Genera un gr√°fico de barras horizontal para visualizar el resultado anterior.

##### Ejercicio 5.2A

```python
# Importamos el DataFrame de pel√≠culas
df_movies = pd.read_csv('./data/imdb_movies.csv')
df_movies.head()
```

```python
# Obtenemos los 10 actores principales con m√°s apariciones en pel√≠culas
top_actors = df_movies['actor_1_name'].value_counts().head(10)
top_actors
```

#### Ejercicio 5.3
---

> Dataset a utilizar: `imdb_movies.csv`

**5.3A:** Crea un histograma para visualizar la distribuci√≥n de la puntuaci√≥n de IMDb ('imdb_score').

**5.3B:** Calcula las estad√≠sticas descriptivas del campo 'imdb_score' y las estad√≠sticas sobre la frecuencia de aparici√≥n del campo 'director_name'.

**5.3C:** Genera un gr√°fico de densidad para comparar la distribuci√≥n de la columna 'imdb_score' para los 5 directores con mayor n√∫mero de pel√≠culas.

##### Ejercicio 5.3A

```python
# Importamos el DataFrame de pel√≠culas
df_movies = pd.read_csv('./data/imdb_movies.csv')
df_movies.dtypes
```

```python
# Importamos Seaborn
import seaborn as sns
```

```python
# Creamos un histograma de 20 barras con la distribuci√≥n del 'imdb_score'
sns.histplot(df_movies['imdb_score'], bins=20)
plt.show()
```

##### Ejercicio 5.3B

```python
# Obtenemos estad√≠sticas descriptivas para 'imdb_score'
df_movies['imdb_score'].describe()
```

```python
# Obtenemos estad√≠sticas sobre la frecuencia para 'director_name'
df_movies['director_name'].describe()
```

##### Ejercicio 5.3C

```python
# Almacenamos en una lista los cinco directores con m√°s pel√≠culas
list_directors = df_movies['director_name'].value_counts().reset_index().head().loc[:,'director_name']
list_directors
```

```python
# Creamos el gr√°fico de densidad del 'imdb_score' dividiendo por director
sns.displot(df_movies[df_movies['director_name'].isin(list_directors)], kind='kde', x='imdb_score', hue='director_name')
plt.show()
```

#### Ejercicio 5.7
---

> Dataset a utilizar: `starbucks_beverages.csv`

Crea un mapa de calor que muestre la matriz de correlaci√≥n entre las diferentes variables de informaci√≥n nutricional de las bebidas de Starbucks.

```python
# Importamos el DataFrame de bebidas de Starbucks
df_starbucks = pd.read_csv('./data/starbucks_beverages.csv')
df_starbucks.head()
```

```python
# Calculamos la matriz de correlaci√≥n de las variables num√©ricas del DataFrame
corr_matrix = df_starbucks.select_dtypes(include='number').corr()
corr_matrix
```

```python
# Creamos un mapa de calor de la matriz de correlaci√≥n
sns.heatmap(corr_matrix, annot=True, cmap='RdBu', annot_kws={"size": 6}, vmin=-1)
plt.show()
```

#### Ejercicio 5.8
---

> Dataset a utilizar: `fortune_1000.csv`

**5.8A:** Crea un gr√°fico de dispersi√≥n entre el cambio en el ranking ('Change in Rank') y los beneficios ('Profits') de las empresas.

**5.8B:** Calcula la correlaci√≥n de Pearson entre el n√∫mero de empleados ('Employees') y la facturaci√≥n ('Revenue').

**5.8C:** Visualiza con un gr√°fico de disperaci√≥n la relaci√≥n entre las dos variables anteriores para las empresas de los sectores: 'Retailing', 'Food, Beverages & Tobacco' y 'Hotels, Restaurants & Leisure'.

##### Ejercicio 5.8A

```python
# Importamos el DataFrame de empresas estadounidenses
df_fortune = pd.read_csv('./data/fortune_1000.csv')
df_fortune.head()
```

```python
# Creamos un gr√°fico de dispersi√≥n entre 'Change in Rank' y 'Profits'
sns.scatterplot(df_fortune, x='Change in Rank', y='Profits')
plt.show()
```

##### Ejercicio 5.8B

```python
# Importamos Scipy
import scipy
```

```python
# Calculamos el correlaci√≥n entre 'Revenue' y 'Employees'
scipy.stats.pearsonr(df_fortune['Revenue'], df_fortune['Employees'])
```

##### Ejercicio 5.8C

```python
# Definimos la condici√≥n para filtrar por los tres sectores espec√≠ficos
cond = df_fortune['Sector'].isin(['Retailing', 'Food, Beverages & Tobacco', 'Hotels, Restaurants & Leisure'])
```

```python
# Creamos un gr√°fico de dispersi√≥n entre 'Revenue' y 'Employees' con los sectores filtrados
sns.relplot(df_fortune[cond], x='Revenue', y='Employees', col='Sector')
plt.show()
```
