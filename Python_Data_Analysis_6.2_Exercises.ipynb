#### Ejercicio 6.1
---

> Dataset a utilizar: `star_wars.csv`

**6.1A:** Crea un gr√°fico de barras que muestre los 10 planetas ('Homeworld') con m√°s personajes en el conjunto de datos.

**6.1B:** Incluye etiquetas en las barras que indiquen el n√∫mero de personajes correspondientes a cada planeta.

##### Ejercicio 6.1A

```python
# Importamos Pandas
import pandas as pd
```

```python
# Importamos el DataFrame de Star Wars
df_starwars = pd.read_csv('./data/star_wars.csv')
df_starwars.head()
```

```python
# Contamos la cantidad de veces que aparece cada planeta y obtenemos los 10 con m√°s filas (personajes)
homeworld_counts = df_starwars['Homeworld'].value_counts().reset_index().head(10)
homeworld_counts
```

```python
# Importamos Matplotlib y Seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Ajustamos el tama√±o del gr√°fico
plt.figure(figsize=(10, 5))

# Creamos un gr√°fico de barras de los 10 pa√≠ses con m√°s personajes
sns.barplot(homeworld_counts, x='Homeworld', y='count', color='grey')
plt.show()
```

##### Ejercicio 6.1B

```python
# Ajustamos el tama√±o del gr√°fico
plt.figure(figsize=(10, 5))

# Creamos un gr√°fico de barras de los 10 pa√≠ses con m√°s personajes
sns.barplot(homeworld_counts, x='Homeworld', y='count', color='grey')

# A√±adimos el t√≠tulo y las etiquetas a los ejes
plt.title('Top 10 Planets with the Most Characters in Star Wars')
plt.xlabel('')
plt.ylabel('Total Characters')

# A√±adimos el n√∫mero de personajes a cada barra
for index, value in enumerate(homeworld_counts['count']):
    plt.text(index, value, f'{value:,.0f}', ha='center', va='bottom', color='grey')

# Eliminamos el borde superior y derecho
sns.despine()

# Mostramos el gr√°fico
plt.show()
```

#### Ejercicio 6.2
---

> Dataset a utilizar: `fortune_1000.csv`

Visualiza con un gr√°fico de barras horizontales el cambio en el ranking de las empresas del sector 'Media', ordenando las barras de mayor a menor seg√∫n el cambio.

```python
# Importamos el DataFrame de empresas estadounidenses
df_fortune = pd.read_csv('./data/fortune_1000.csv')
df_fortune.head()
```

```python
# Preparamos el DataFrame filtrando las empresas del sector 'Media' y orden√°ndolo por 'Change in Rank'
df_fortune = df_fortune[df_fortune['Sector'] == 'Media']
df_fortune = df_fortune.sort_values('Change in Rank', ascending=False)

df_fortune.head(10)
```

```python
# Ajustamos el tama√±o del gr√°fico
plt.figure(figsize=(10, 7.5))

# Creamos el gr√°fico de barras horizontales para visualizar la columna 'Change in Rank'
sns.barplot(data=df_fortune, x='Change in Rank', y='Company', color='grey')

# Mostramos el gr√°fico
plt.show()
```

```python
# Creamos una nueva columna para categorizar los valores como 'Positive' o 'Negative' en funci√≥n de 'Change in Rank'
df_fortune['Color'] = df_fortune['Change in Rank'].apply(lambda x: 'Positive' if x > 0 else 'Negative')

# Ajustamos el tama√±o del gr√°fico
plt.figure(figsize=(10, 7.5))

# Creamos el gr√°fico de barras con los colores personalizados
sns.barplot(data=df_fortune, x='Change in Rank', y='Company', palette={'Positive':'lightskyblue', 'Negative':'lightcoral'}, hue='Color', legend=False)

# A√±adimos el t√≠tulo
plt.title('YoY Change in Rank of Media Companies')
plt.ylabel('')

# Eliminamos el borde superior y derecho
sns.despine()

# Mostramos el gr√°fico
plt.show()
```

#### Ejercicio 6.3
---

> Dataset a utilizar: `amzn_stock.csv`

Crea un gr√°fico que muestre la media de cotizaci√≥n del cierre semanal de las acciones de Amazon en 2020, incluyendo una l√≠nea vertical en la semana 12 para se√±alar la fecha del inicio del confinamiento en Estados Unidos debido a la pandemia del COVID-19.

```python
# Importamos el DataFrame con las cotizaciones de Amazon
df_amzn = pd.read_csv('./data/amzn_stock.csv')
df_amzn.head()
```

```python
# Convertimos la columna 'Date' a tipo fecha
df_amzn['Date'] = pd.to_datetime(df_amzn['Date'])

# Creamos una columna con el n√∫mero de semana
df_amzn['Week Number'] = df_amzn['Date'].dt.isocalendar().week
```

```python
# Creamos las condiciones para seleccionar las cotizaciones del 2020
cond_1 = df_amzn['Date'] >= '2020-01-01'
cond_2 = df_amzn['Date'] < '2020-12-31'

# Aplicamos las dos condiciones
amzn_2020 = df_amzn[cond_1 & cond_2]
amzn_2020.head()
```

```python
# Agrupamos los datos por semana y calculamos la media de 'Close'
amzn_weekly = amzn_2020.groupby('Week Number').agg({'Close':'mean'}).reset_index()
amzn_weekly.head()
```

```python
# Ajustamos el tama√±o del gr√°fico
plt.figure(figsize=(10, 5))

# Creamos un gr√°fico de l√≠neas con la evoluci√≥n de 'Close'
sns.lineplot(amzn_weekly, x='Week Number', y='Close', color='grey')

# A√±adimos el t√≠tulo y las etiquetas a los ejes
plt.title('Amazon Stock Price Evolution')
plt.xlabel('Week Number')
plt.ylabel('Close Price (USD)')

# Ajustamos el l√≠mite del eje y
plt.ylim(0)

# A√±adimos una l√≠nea discontinua en la semana 12
plt.axvline(x=12, color='grey', linestyle='--', linewidth=1)

# Eliminamos el borde superior y derecho
sns.despine()

# Mostramos el gr√°fico
plt.show()
```

#### Ejercicio 6.8
---

> Dataset a utilizar: `amzn_stock.csv`

Visualiza el volumen medio hist√≥rico de transacciones semanales de Amazon utilizando un mapa de calor.

```python
# Importamos el DataFrame con las cotizaciones de Amazon
df_amzn = pd.read_csv('./data/amzn_stock.csv')
df_amzn.head()
```

```python
# Convertimos la columna 'Date' a tipo fecha
df_amzn['Date'] = pd.to_datetime(df_amzn['Date'])
```

```python
# Creamos una columna con el a√±o
df_amzn['Year'] = df_amzn['Date'].dt.isocalendar().year

# Creamos una columna con el n√∫mero de semana
df_amzn['Week Number'] = df_amzn['Date'].dt.isocalendar().week
```

```python
# Agrupamos por a√±o y semana, y calculamos el volumen de transacciones promedio
amzn_weekly = df_amzn.groupby(['Year', 'Week Number']).agg({'Volume':'mean'})
amzn_weekly
```

```python
# Pivotamos el DataFrame para que cada fila sea un a√±o, cada columna una semana, y los valores sean los valores de 'Volume'
df_amzn_pivoted = amzn_weekly.pivot_table(index='Year', columns='Week Number', values='Volume')
df_amzn_pivoted.head()
```

```python
# Ajustamos el tama√±o del gr√°fico
plt.figure(figsize=(20, 10))

# Creamos un mapa de calor para representar el 'Volume' por semana y a√±o
sns.heatmap(df_amzn_pivoted, xticklabels=True, square=True, cbar=False, cmap='OrRd', linecolor='white', linewidths=0.5)

# A√±adimos el t√≠tulo al mapa de calor
plt.title('Average Weekly Trading Volume of Amazon\n', fontsize=16, fontweight='bold')
plt.ylabel('')

# Mostramos el gr√°fico
plt.show()
```

#### Ejercicio 6.9

> Dataset a utilizar: `laliga_results.csv`

Crea un mapa de calor donde las filas correspondan a los 20 equipos de LaLiga y las columnas representen los goles de cada jornada. El valor de cada celda debe indicar el n√∫mero de goles anotados por cada equipo en cada jornada.

```python
# Importamos el DataFrame con los resultados de LaLiga
df_laliga = pd.read_csv('./data/laliga_results.csv')
df_laliga.head()
```

```python
# Creamos la columna 'Matchday ' en funci√≥n del √≠ndice para que cada 10 filas correspondan a una jornada
df_laliga['Matchday'] = 'M' + ((df_laliga.index // 10 + 1).astype(str).str.zfill(2)) # Aseguramos con '.str.zfill(2)' que el n√∫mero de la jornada tenga dos d√≠gitos
df_laliga.head(20)
```

```python
# Creamos un DataFrame con los goles marcados por los equipos locales
df_laliga_home = df_laliga[['Matchday', 'Home Team', 'HT Goals']]

# Creamos un DataFrame con los marcados por los equipos visitantes
df_laliga_away = df_laliga[['Matchday', 'Away Team', 'AT Goals']]
```

```python
# Renombramos los nombres de columnas de los dos DataFrames para que sean iguales
df_laliga_home = df_laliga_home.rename(columns = {'Home Team':'Club', 'HT Goals':'Goals'})
df_laliga_away = df_laliga_away.rename(columns = {'Away Team':'Club', 'AT Goals':'Goals'})
```

```python
# Unimos los DataFrames por sus filas
df_laliga_goals = pd.concat([df_laliga_home, df_laliga_away])
df_laliga_goals
```

```python
# Pivotamos el DataFrame para que cada fila corresponda a un 'Club', cada columna a un 'Matchday' y cada celda contenga el valor de 'Goals'
df_laliga_pivoted = df_laliga_goals.pivot_table(index='Club', columns='Matchday', values='Goals', fill_value=0)

# Ordenamos las columnas con la funci√≥n 'sorted()' para que las jornadas est√©n en orden cronol√≥gico
df_laliga_pivoted = df_laliga_pivoted[sorted(df_laliga_pivoted.columns)]
df_laliga_pivoted
```

```python
# Ajustamos el tama√±o del gr√°fico
plt.figure(figsize=(10, 5))

# Creamos el mapa de calor para visualizar los goles de cada equipo en cada jornada
sns.heatmap(df_laliga_pivoted, xticklabels=False, square=True, cbar=False, cmap='OrRd', linecolor='white', linewidths=0.5)

# A√±adimos el t√≠tulo al gr√°fico y las etiquetas a los ejes
plt.title('Matchday Goals Heatmap in LaLiga 2023-2024')
plt.ylabel('Matchdays')
plt.ylabel('')

# Mostramos el gr√°fico
plt.show()
```
